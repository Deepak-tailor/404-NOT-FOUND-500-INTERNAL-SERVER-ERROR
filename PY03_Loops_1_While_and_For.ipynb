```python
test_cases = int(input("How many executions you want? "))
t = 1
while(t<=test_cases):
  number = int(input("Please enter a number to generate the table: "))
  i =1
  print("The multiplication table of ",number)
  while(i<=10):
    print(i*number,end=" ")
    i=i+1
  print("\n")
  t=t+1
```
